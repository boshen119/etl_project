```python
import os
import pandas as pd
import csv
```

```python
# Load scraped data csv to Python
master_sd_df = pd.read_csv("slickdeals.csv")
print(master_sd_df)
```

```python
# Make a copy of dataframe
slick_items_df = master_sd_df
```

```python
# Adding website to dataframe
slick_items_df["Store"] = "slickdeals.net"
slick_items_df
```

```python
# Split item column after "free"

# leaves "free shipping" and "store pick up"
pick_up_method = pd.DataFrame(slick_items_df.Item.str.split('+').str[1])

# Uniform 'Free Shipping'
string = 'Free S&amp;H'
string2 = 'Free S/H'

for index, row in pick_up_method.iterrows():
    string.replace('Free S&amp;H', 'Free Shipping')
pick_up_method

for index, row in pick_up_method.iterrows():
    string2.replace ('Free S/H', 'Free Shipping')
    
# Store to CSV
pick_up_method.to_csv('slickdeal_items_w_pickup.csv', index=True)
```

```python
#drops the shipping/store pickup
split_items = pd.DataFrame(slick_items_df.Item.str.split('+').str[0])


# split_items = pd.DataFrame(slick_items_df.Item.str.split('&amp;').str[0])
split_items.to_csv("slickdeal_items.csv", index=True)
```

```python
merged_items_df = split_items.merge(pick_up_method, left_index=True, right_index=True, how='inner')
```

```python
merged_items_df = merged_items_df.rename(columns = {"Item_x" : "Item",
                                  "Item_y" : "Pick Up Notes"})
merged_items_df.to_csv('merged_item_columns.csv', index=True)
```

```python
split_items = pd.DataFrame(slick_items_df.Item.str.split('+',1))
split_items
```

```python
final_df = pd.merge(slick_items_df, merged_items_df, left_index=True, right_index=True)
final_df
```

```python
final_df = final_df.rename(columns = {"Item_y" : "Item"})
final_df
```

```python
final_df = final_df.drop(['Unnamed: 0'], 1)
final_df
```

```python
# Reorder columns with data
final_df = final_df[['Item', 'Old Price', 'Sale Price', 'Store', 'Pick Up Notes']]
```
